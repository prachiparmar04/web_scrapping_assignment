## 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

        - AS the Front Page does not contain much Reviews, i am scrapping information from - https://bookpage.com/reviews


```python
request_book = requests.get('https://bookpage.com/reviews')
```

```python
request_book
```

```python
soup_book = BeautifulSoup(request_book.content)
```

```python
booknames = soup_book.find_all('h4',class_='italic')
```

```python
booknames
```

```python
authornames = soup_book.find_all('p',class_='sans bold')
```

```python
authornames
```

```python
genres = soup_book.find_all('p',class_='genre-links hidden-phone')
```

```python
genres
```

```python
book_names = []

for i in range(0,5):
    count = 0
    if(len(booknames[i].find('a'))==1):
        book_names.append(booknames[i].find('a').text)
    else:
        for j in booknames[i].find('a'):
            
            count = count+1
            if(count == 3):
                book_names.append(j)
```

```python
book_names
```

```python
authors = []
for i in authornames[:5]:
    authors.append(i.text.replace('\n',''))
```

```python
authors
```

```python
genres_names = []
for i in genres[:5]:
   # for j in i.find('a'):
    genres_names.append(i.text.replace('\n',''))
```

```python
genres_names
```

```python
reviews = soup_book.find_all('div',class_='flex-article-content')
```

```python
reviews
```

#### Observations:
    - As Reviews_book for each Book Section does belong to one Common Html Tag. Some Review belong to <p></p> and some belong to <p class='excerpt></p>
    - So we need to make a customize code for extracting both type of TAGS.

```python
reviews_books = []

for i in range(0,5):
    
    if(len(reviews[i].find_all('p'))==3):
        for j in reviews[i].find_all('p',class_='excerpt'):
            reviews_books.append(j.text.replace('\n',''))
    else:
         
        if(len(reviews[i].find_all('p'))==4):
            c = 0
            for j in reviews[i].find_all('p'):
                c=c+1
                if(c==4):
                     reviews_books.append(j.text.replace('\n',''))
      
```

```python
reviews_books
```

```python
books_dict = {
    'Book Name' : book_names,  'Author Name': authors, 'Genres': genres_names, 'Review': reviews_books
    
}
```

```python
book_Table = pd.DataFrame(books_dict)
```

```python
book_Table
```

```python
book_Table.set_index('Book Name')
```
